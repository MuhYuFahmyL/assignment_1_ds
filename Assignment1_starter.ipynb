## Histogram

A histogram is a way of representing the frequency distribution of numeric dataset. The way it works is it partitions the x-axis into bins, assigns each data point in our dataset to a bin, and then counts the number of data points that have been assigned to each bin. So the y-axis is the frequency or the number of data points in each bin. Note that we can change the bin size and usually one needs to tweak it so that the distribution is displayed nicely.

**Question:**

Berapa distribusi frekuensi jumlah kasus dari berbagai kasus yang terjadi pada tahun 2016 di london

```python
criminal_in_2016 = df[df['value'] != 0]
criminal_in_2016 = criminal_in_2016[['lsoa_code','major_category','value','month','year']]
criminal_in_2016.set_index('lsoa_code', inplace=True)
criminal_in_2016 = pd.crosstab(criminal_in_2016['year'],criminal_in_2016['major_category'])
criminal_in_2016 = criminal_in_2016.loc[[2016],:].T
criminal_in_2016

count, bin_edges = np.histogram(criminal_in_2016)
criminal_in_2016.plot(kind='hist',figsize=(8,5),xticks=bin_edges,yticks=count)
plt.title('Histogram of kind of cases in london 2016') # add a title to the histogram
plt.ylabel('Number of major category') # add y-label
plt.xlabel('Number of total cases') # add x-label

plt.show()
```

**Insight:**

pada sumbu x menunjukan kisaran populasi imigran dalam interval 129159. sumbu y mewakili jumlah kasus yang terjadi

## Bar Charts (Dataframe) <a id="10"></a>

A bar plot is a way of representing data where the *length* of the bars represents the magnitude/size of the feature/variable. Bar graphs usually represent numerical and categorical variables grouped in intervals. 

To create a bar plot, we can pass one of two arguments via `kind` parameter in `plot()`:

* `kind=bar` creates a *vertical* bar plot
* `kind=barh` creates a *horizontal* bar plot

**Question:**

Pada tahun berapa terjadi kasus pencurian terbanyak?

```python
criminal_in_theft = df[df['value'] != 0]
criminal_in_theft = criminal_in_theft[['lsoa_code','major_category','value','month','year']]
criminal_in_theft.set_index('lsoa_code', inplace=True)
criminal_in_theft = pd.crosstab(criminal_in_theft['year'],criminal_in_theft['major_category']).T
criminal_in_theft = criminal_in_theft.loc[['Theft and Handling'],:].T
criminal_in_theft.plot(kind='bar',figsize=(10,6))
plt.xlabel('Year') # add to x-label to the plot
plt.ylabel('Number of Cases') # add y-label to the plot
plt.title('Theft and Handling cases from 2008 to 2016 in London') # add title to the plot

plt.show()
```

**Insight:**

Berdasarkan hasil pengolahan data yang divisualisasikan pada Bar Chart diatas, dapat dilihat tahun 2012 menjadi tahun dengan kejadian pencurian paling tinggi

A `pie chart` is a circualr graphic that displays numeric proportions by dividing a circle (or pie) into proportional slices. ## Box Plots <a id="8"></a>

A `box plot` is a way of statistically representing the *distribution* of the data through five main dimensions: 

- **Minimun:** Smallest number in the dataset.
- **First quartile:** Middle number between the `minimum` and the `median`.
- **Second quartile (Median):** Middle number of the (sorted) dataset.
- **Third quartile:** Middle number between `median` and `maximum`.
- **Maximum:** Highest number in the dataset.

**Question:**

Bagaimana Perbandingan angka distribusi untuk kejahatan Perampokan dan Kerusakan Kriminal selama 2008 hingga 2016 

```python
criminal_in_burglary_criminal_damage = df[df['value'] != 0]
criminal_in_burglary_criminal_damage = criminal_in_burglary_criminal_damage[['lsoa_code','major_category','value','month','year']]
criminal_in_burglary_criminal_damage.set_index('lsoa_code', inplace=True)
criminal_in_burglary_criminal_damage = pd.crosstab(criminal_in_burglary_criminal_damage['year'],criminal_in_burglary_criminal_damage['major_category']).T
criminal_in_burglary_criminal_damage = criminal_in_burglary_criminal_damage.loc[['Burglary','Criminal Damage'],:].T
criminal_in_burglary_criminal_damage.plot(kind='box',figsize=(8,6))

plt.title('Box Plot of Criminal Damage and Burglary')
plt.ylabel('yLabel')

plt.show()
criminal_in_burglary_criminal_damage.describe()
```

## Scatter Plots <a id="10"></a>

A `scatter plot` (2D) is a useful method of comparing variables against each other. `Scatter` plots look similar to `line plots` in that they both map independent and dependent variables on a 2D graph. While the datapoints are connected together by a line in a line plot, they are not connected in a scatter plot. The data in a scatter plot is considered to express a trend. With further analysis using tools like regression, we can mathematically calculate this relationship and use it to predict trends outside the dataset.

**Question:**

Bagaimana tren angka kriminal di kota london selama 2008 hingga 2016

```python
criminal_all = df[df['value'] != 0]
criminal_all = criminal_all[['major_category','value','year']]
criminal_all = criminal_all.groupby(['year']).sum()
criminal_all.index = map(int,criminal_all.index)
criminal_all.reset_index(inplace=True)
criminal_all.columns = ['year','total']
criminal_all.plot(kind='scatter', x='year', y='total', figsize=(10, 6), color='darkblue')
plt.title('Total Cases in London from 2008 to 2016')
plt.xlabel('Year')
plt.ylabel('Number of Cases')
plt.show()
```

**Insight:**

Kita dapar mengamati bagaimana scatter plot tidak menghubungkan antar data points. kita dapat dengan jelas mengamati tren angka kriminalitas dan terlihat pada tahun 2014 menjadi tahun dengan angka kriminal terendah selama dari 2008 hingga 2016
